### Part b. How many malignant cases are there in total?

Based on the cancer dataset description, "Class Distribution: 212 - Malignant, 357 - Benign". 

```python
# Add classifier to dataframe to identify those with malignant or benign classes.

df["cancer_type"] = cancer.target
new_label = {"cancer_type": {0: "malignant", 1: "benign"}}
df.replace(new_label , inplace = True)
```

### Part c. Make a scatterplot of ‘mean compactness’ and ‘mean perimeter’. 

```python
# Make a scatterplot of ‘mean compactness’ and ‘mean perimeter’. Try to color the points by their label (i.e. malignant or benign). 

sns.jointplot(x = "mean compactness", y = "mean perimeter", data = df, hue = "cancer_type")
plt.title("Scatterplot of ‘mean compactness’ and ‘mean perimeter’", y = 1.25)
```

```python
#  Make pairwise scatterplots of ‘mean texture’, ‘mean perimeter’, ‘mean compactness’, and ‘mean radius’.
sns.pairplot(df, vars=['mean texture', 'mean perimeter', 'mean compactness', 'mean radius'], hue='cancer_type')
plt.title("Pair scatterplots of ‘mean texture’, ‘mean perimeter’, ‘mean compactness’, and ‘mean radius’.", y = 4.25, x = -1)
```

### Part d. Do you notice that two of these features seem to be particularly correlated? Why might this be?

The two features that seem to be best correlated are mean radius and mean perimeter. This is likely directly related to each other's measurements due to the nature of how perimeter is calculated by the radius of the cancer ($P = 2 \pi r$).
